# 소개정보조회

```python
raw_data = list()
```

```python
for i, contentId in enumerate(df["contentid"][1996:], 1996):
	response = requests.get("https://apis.data.go.kr/B551011/KorService2/detailIntro2", {"MobileOS": "ETC", "MobileApp": "Python", "_type": "json", "contentId": contentId, "contentTypeId": "15", "serviceKey": serviceKey_decoding}).json()
	if response.get("response").get("body").get("totalCount"):
		item = response.get("response").get("body").get("items")
		raw_data.append(item)
	time.sleep(2)
	print(i, f"contentId: {contentId}")
```

```python
pd.DataFrame([a["item"][0] for a in raw_data]).to_csv("data/소개정보조회_2025_11_15.csv", index=False)
fuck = [a["item"][0] for a in raw_data]
with open("data/소개정보조회_2025_11_15.json", "w", encoding="utf-8") as fa:
	json.dump(fuck, fa, ensure_ascii=False, indent="\t")
```
