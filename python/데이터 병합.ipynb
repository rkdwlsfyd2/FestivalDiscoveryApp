```python
import json
import pandas as pd
```

# 1. 데이터 로드

```python
with open(f"data/공통정보조회_final.json", encoding="utf-8") as f:
	common_info = json.load(f)

with open(f"data/소개정보조회_final.json", encoding="utf-8") as f:
	detail_intro = json.load(f)

with open(f"data/반복정보조회_final.json", encoding="utf-8") as f:
	detail_info = json.load(f)

with open(f"data/이미지정보조회_final.json", encoding="utf-8") as f:
	image_info = json.load(f)
```

```python
common_info_df = pd.DataFrame(common_info)
detail_intro_df = pd.DataFrame(detail_intro)
detail_info_df = pd.DataFrame(detail_info)
image_info_df = pd.DataFrame(image_info)
```

# 2. 축제 소개, 내용 병합

### 2-1. 사용하지 않는 컬럼 제거

```python
detail_info_df = detail_info_df[["contentid", "infotext"]]
detail_info_df.head()
```

### 2-2. NULL 확인

```python
detail_info_df.info()
```

### 2-4. contentid로 그룹화하여 축제 소개, 내용 병합

```python
detail_info_df = detail_info_df.groupby("contentid").agg(" ".join)
detail_info_df.head()
```

# 3. json 파일로 저장

```python
with open("data/infotext_list_final.json", "w", encoding="utf-8") as fw:
    json.dump(detail_info_df["infotext"].to_list(), fw, ensure_ascii=False, indent="\t")
```
